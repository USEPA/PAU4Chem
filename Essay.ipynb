```python
import pandas as pd
import numpy as np
from itertools import combinations 
```

```python
def division(row):
    cases = {}
    n_cases = 1
    elements_total = ['OFF-SITE - TOTAL TRANSFERRED FOR FURTHER WASTE MANAGEMENT', 
                      'OFF-SITE - TOTAL POTW TRANSFER',
                      'ON-SITE - TOTAL RELEASES', 
                      'ON-SITE - ENERGY RECOVERY']
    for n_elements_sum in [0, 1, 2, 3, 4]:
        comb = combinations(elements_total, n_elements_sum)
        for columns in comb:
            if n_elements_sum == 0:
                sumatory = 0.0
                fugitive = 0.0
            else:
                sumatory = sum(row[list(columns)])
                if 'ON-SITE - TOTAL RELEASES' in list(columns):
                    fugitive = row['ON-SITE - FUGITIVE AIR EMISSIONS']
                else:
                    fugitive = 0.0
            try:
                cases.update({n_cases: row['ON-SITE - RECYCLED']/(row['ON-SITE - RECYCLED'] + row['ON-SITE - TREATED'] + sumatory - fugitive)*100})
            except ZeroDivisionError:
                cases.update({n_cases: np.nan})
            n_cases = n_cases + 1
    return cases
```

```python
columns = ['ON-SITE - TREATED', 'OFF-SITE - TOTAL TRANSFERRED FOR FURTHER WASTE MANAGEMENT',
          'OFF-SITE - TOTAL POTW TRANSFER', 'ON-SITE - ENERGY RECOVERY', 'ON-SITE - TOTAL RELEASES', 
          'TRIFID', 'CHEMICAL NAME', 'ON-SITE - RECYCLED', 'ON-SITE - FUGITIVE AIR EMISSIONS', 
          'ON-SITE - ENERGY RECOVERY']
```

```python
df = pd.read_csv(Path_TRI, usecols = columns, low_memory = True)
```

```python
df.info()
df = df.loc[(df['ON-SITE - RECYCLED'] != 0.0) & pd.notnull(df['ON-SITE - RECYCLED'])]
```

```python
df['RECYCLING EFFICIENCY'] = df.apply(lambda x: division(x), 
                                       axis = 1)
```
